```python
from numpy import random
def init_poids(dimEntree, dimSortie):
    random.seed(2)
    poids = None
    import numpy as np
    poids= np.random.uniform(-50, 50, size = (dimEntree,dimSortie))
    return poids
```

```python
## Test du réseau de neurones
```

```python
import numpy as np

def calcul_neurone(j, e, poids):
    resultat_j=0
    mult = e*poids[:,j]
    for i in mult:
        resultat_j+=i
    return resultat_j
```

```python
print(calcul_neurone(3, d[9][0], poids))
print(calcul_neurone(9, d[9][0], poids))
```

```python
def calcul_reseau(e, poids):
    res=0
    resultat = []
    for j in range(10):
        mult = e*poids[:,j]
        for i in mult:
            res+=i
        resultat.append(res)
        
    return resultat
```

```python
calcul_reseau(d[9][0], poids)
```

```python
## Apprentissage
```

```python
def apprendre_neurone(e, poids, j, sortie_attendue):
    h = 5
    valeur_calculee = 0
    valeur_calculee = calcul_neurone(j, e, poids)
    for i in e: 
        valeur_desiree = 0
        if sortie_attendue == j:
            valeur_desiree = 1
        else:
            valeur_desiree = 0
        poids[i][j] = poids[i][j] + (valeur_desiree - valeur_calculee) * e[i] * h
    return poids    
```

```python
apprendre_neurone(d[9][0], poids, 8, 0)
```

```python
def apprendre_reseau(e, poids, sortie_attendue):
    for j in range(10):
        apprendre_neurone(e, poids, j, sortie_attendue)
        return poids
```

```python
apprendre_reseau(d[9][0], poids, 1)
```

```python
def apprendre(d, poids):
    for j in d:
        for i in d[j]:
            apprendre_reseau(i, poids, j)  
            return poids
```

```python
apprendre(d, poids)
```
